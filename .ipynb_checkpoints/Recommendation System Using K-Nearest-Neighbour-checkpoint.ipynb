```python
#checking for NULL values in Movies Dataset.
moviesCollumns = movies.columns
for i in moviesCollumns:
    print(movies[i].isnull().sum())
```

```python
#checking for NULL values in ratings dataset.
ratingsCollumns = ratings.columns
for i in ratingsCollumns:
    print(ratings[i].isnull().sum())
```

```python
# making list of movies that are common in both of the datasets.
for i in range(0, 10):
    print(movies['movieId'].loc[i],"", ratings['movieId'].loc[i])
```

```python
ratings.head()
```

```python
dataset = pd.merge(movies, ratings, on='movieId')
dataset.head(20)
```

```python
# for i in dataset['movieId']:
#     print(dataset['userId'].loc[i].sum())
dataset['userId'].loc[7].sum()
```

```python
movieRatingCount = dataset.dropna(subset=['title'])
movieRatingCount = dataset.groupby(by='title').count().reset_index().rename(columns={'rating': 'totalRatingCount'})[['title', 'totalRatingCount']]
movieRatingCount.head(10)
```

```python
datasetWithRatingCount = pd.merge(dataset, movieRatingCount, on='title')
datasetWithRatingCount.head()
```

```python
datasetWithRatingCount.info()
```

```python
dfColumns = datasetWithRatingCount.columns
for i in dfColumns:
    print(datasetWithRatingCount[i].isnull().sum())
```

```python
#Only recommend the movies with total rating count greater than or equal to popularity threshold.
popularityThreshold = 50
ratingPopularMovies = datasetWithRatingCount.query('totalRatingCount >= @popularityThreshold')
ratingPopularMovies.head()
```

```python
#In the tutorial they dropped genres and timestamp... I mean timestap is useless but why remove generes...like why...
ratingPopularMovies.columns
```

```python
#making the pivot matrix. Rows = movie titles, Collumns = user Ids, values = ratings
movies_features_df = ratingPopularMovies.pivot_table(index='title', columns='userId', values='rating').fillna(0)
movies_features_df.head()
```

```python
import sklearn, scipy
```

```python
from scipy.sparse import csr_matrix
movieFeatureMatrix = csr_matrix(movies_features_df.values)
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movieFeatureMatrix)
```

```python
movieFeatureMatrix.shape
```

```python
query_index = np.random.choice(movies_features_df.shape[0])
print(query_index)
distance, indices = model_knn.kneighbors(movies_features_df.iloc[query_index, :].values.reshape(1,-1), n_neighbors = 6)
```
